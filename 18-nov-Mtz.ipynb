```python
def parabola(a, b, c ):
    y = []
    x = []
    for i in range(-10, 11):
        x0 = i
        y0 = a*x0**2 + b*x0 + c
        x.append(x0)
        y.append(y0)
    return x, y
```

```python
x, y = parabola(-1, 0, 0)
plt.plot(x,y)
```

```python
def f_planet(x0, vx0, y0, z0, t0=0, dt=0.01 ,steps=100):
    G = 4*(mat.pi**2)
    M = 1
    rp = mat.sqrt( x0**2 + y0**2 + z0**2 ) 
    x = [x0] 
    t = [t0]
    for i in range(1, steps):
        Vxi = vx0 - ( (G*M) / rp**3 )*x0 *dt 
        xi = x0 + Vxi*dt
        x.append(xi)
        t.append(t0+dt*i)
        x0 = xi
        vx0 = Vxi 
    return x, t
```

```python
#Merc
x, t = f_planet(0.0508011, 0.022318, -0.4011642, -0.2195576, steps=25
               )
plt.plot(t , x)
```

```python
#Sat
x, t = f_planet(-3.743, -0.005, 7.568, 3.287, steps=1100) 
plt.plot(t, x)
```

```python
def f_planet1( x0, y0, z0, vx0, vy0, vz0, t0=0, dt=0.01, steps = 100 ):
    G = 4 * ( mat.pi** 2)
    M = 1
    rp = mat.sqrt( x0**2 + y0**2 + z0**2 ) 
    x = [ x0 ] 
    y = [ y0 ]
    z = [ z0 ]
    t = [ t0 ]
    for i in range( 1, steps ):
        Vxi = vx0 - ( (G*M) / rp**3 )*x0 *dt 
        xi = x0 + Vxi*dt
        Vyi = vy0 - ( (G*M) / rp**3 )*y0 *dt 
        yi = y0 + Vyi*dt
        Vzi = vz0 - ( (G*M) / rp**3 )*z0 *dt 
        zi = z0 + Vzi*dt
        x.append( xi )
        y.append( yi )
        z.append( zi )
        t.append( t0+dt*i )
        x0 = xi
        vx0 = Vxi 
        y0= yi
        vy0 = Vyi
        z0 = zi
        vz0 = Vzi
        
    return x, y, z, t
```

```python
x, y, z, t = f_planet(0.0508011, 0.022318, -0.4011642, -0.2195576, steps=100)
plt.plot(t , x)
```
